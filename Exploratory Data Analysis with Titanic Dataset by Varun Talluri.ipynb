```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d = pd.read_csv('D:/Code Tech IT Internship/titanic.csv')
print(d.head())
print('\n', d.describe())
print('\n', d.isnull().sum())
```

```python
d.hist(figsize = (12, 8))
plt.title(' THE FOLLOWING ARE THE HISTOGRAMS FOR THE NUMERICAL FEATURES ')
plt.show()
```

```python
n = d.select_dtypes(include = [np.number])
for c in n:
    plt.figure(figsize = (8, 6))
    sns.boxplot(x = d[c])
    plt.title(f' THE BOX PLOT OF {c} IS')
    plt.show()
```

```python
c = n.corr()
plt.figure(figsize = (10, 8))
sns.heatmap(c, annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title(' THE CORRELATION HEATMAP IS ')
plt.show()
```
